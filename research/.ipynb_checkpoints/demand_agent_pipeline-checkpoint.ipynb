```python
import os
import zipfile
import pandas as pd
import numpy as np

# 1. Authenticate with Kaggle API
from kaggle.api.kaggle_api_extended import KaggleApi
api = KaggleApi()
api.authenticate()

# 2. Download and unzip the specific dataset
download_path = '../data/raw/'
os.makedirs(download_path, exist_ok=True)
api.dataset_download_files('raminhuseyn/demand-forecasting-dataset', path=download_path, unzip=True)

# 3. Load data into a Pandas DataFrame
# Note: Check the exact unzipped filename. It is typically 'train.csv' or similar.
csv_file = os.path.join(download_path, 'train.csv')
df = pd.read_csv(csv_file)

# Parse dates to enable sequential sorting
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(by='date').reset_index(drop=True)

print("Dataset Loaded Successfully!")
display(df.head())
```
